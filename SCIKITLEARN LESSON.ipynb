```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np

#import data
heart_disease  = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
#create out feature( x data) and our label( y data)

x = heart_disease.drop("target", axis=1)
y = heart_disease["target"]
```

```python
#import models and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

#default hyperparameters
clf.get_params()
```

```python
#split data 

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
#train/fit the model with the data

clf.fit(x_train,y_train);
```

```python
#make predictions with data set
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
# Evaluate the model on training data 
clf.score(x_train, y_train)
```

```python
#Evaluate model on test data
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
#Improve model accuracy by twirking the n_estimator hyperparameter
np.random.seed(40)

for i in range(10,100,10):
    print(f"Trying model with {i} n-estimator...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test set : {clf.score(x_test, y_test) * 100}")
    print("")
```

```python
#saving our model
import pickle
pickle.dump(clf, open("random-forest-model.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random-forest-model.pkl", "rb"))
loaded_model.score(x_test, y_test)
```

### Filling missing data Using Sklearn

```python
car_sales_missing = pd.read_csv("car-sales-extended-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing.isna().sum()
```

```python
#drop all rows with misiing price
car_sales_missing.dropna(subset=["Price"], inplace=True)
car_sales_missing.isna().sum()
```

```python
#Split data
X = car_sales_missing.drop("Price", axis=1)
y = car_sales_missing["Price"]
```

#### Sklearn way of filling missing data

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

#define values to fill missing data
cat_imputer = SimpleImputer(strategy="constant", fill_value="missing")
door_imputer = SimpleImputer(strategy="constant", fill_value=4)
num_imputer = SimpleImputer(strategy="mean")

#define columns to apply each imputation
cat_features =["Make", "Colour"]
door_feature = ["Doors"]
num_feature = ["Odometer (KM)"]

#map them to their columns using ColumnTransformer
imputer = ColumnTransformer([
    ("cat_imputer", cat_imputer, cat_features),
    ("door_imputer", door_imputer, door_feature),
    ("num_imputer", num_imputer, num_feature)
])

#tranform 
filled_X = imputer.fit_transform(X)
filled_X
```

```python
car_sales_filled = pd.DataFrame(filled_X,
                                columns=["Make","Colour","Doors", "Odometer (KM)"])
car_sales_filled
```

```python
car_sales_filled.isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
#why doors? Doors is categorical but referenced by numbers
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", 
                                   one_hot,
                                   categorical_features)],
                                   remainder="passthrough")
#new X data
transformed_X = transformer.fit_transform(car_sales_filled)
transformed_X
```

```python
#Split data
X = car_sales_missing.drop("Price", axis=1)
y = car_sales_missing["Price"]
y
```

```python
# we now have a clean(no missing values) and all-numeric data, we can fit into a 
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(transformed_X,
                                                    y, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

## Choosing the right estimator/algorithm for our data

or refer to the sklearn documentation for the image cheat sheat at https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
housing_df = pd.DataFrame(housing["data"], columns=housing["feature_names"])
housing_df
```

```python
housing_df["MedHouseVal"] = housing["target"]
housing_df
```

```python
housing_df["target"] = housing["target"]
housing_df
```

```python
housing_df = housing_df.drop("MedHouseVal", axis=1)
housing_df
```

```python
#import estimator
from sklearn.linear_model import Ridge

#set up random seed
np.random.seed(42)

X = housing_df.drop("target", axis=1)
y = housing_df["target"] #per $100,000

#split my data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#call/initialize model
model = Ridge()
model.fit(X_train, y_train)

#check model score
model.score(X_test, y_test)
```
