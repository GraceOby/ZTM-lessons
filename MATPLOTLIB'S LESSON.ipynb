## Matplotlib Workflow

```python
#1. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#2.Create your data(either using an aready numpy, panda dataset or a simple python)
x= [1,2,3,4,5]
y = [100,200,300,400,500]

#3. Create your canvas for plot
fig, ax = plt.subplots(figsize=(10,10))

#4. plot your data on the figure
ax.plot(x,y);

#5. Set styles, ie customize your plot
ax.set(title="Simple plot",
       xlabel= "x-axis",
      ylabel= "y-axis")

#6 Save figure and export
fig.savefig("Simple-plot.png")
```

### 1st Method of creating subplots

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(products.keys(),products.values());
ax4.hist(np.random.randn(100));
```

#### 2nd method of creating subplots

```python
fig, ax =plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5));
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(products.keys(),products.values());
ax[1,1].hist(np.random.randn(100));
```

### Plotting from Pandas DataFrames using matplotlip object-Oriented

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
#using the oop method mixed with pyplot
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
ax.set_xlim(45, 100);
```

```python
over_50.head()
```

```python
#plt.style.use("ggplot")
```

```python
#Using pure OOP
fig, ((ax1,ax2)) = plt.subplots(nrows=2, ncols=1,figsize=(10,10), sharex=True)
scatter1 = ax1.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
#customizing
ax1.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol level")

ax1.legend(*scatter1.legend_elements(), title="Target")
#add horizontal line

ax1.axhline(y=over_50["chol"].mean(),
          linestyle="--")



#for the second axes
scatter2 = ax2.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
#customizing
ax2.set(title="Heart Disease and Thalach",
       xlabel="Age",
       ylabel="Max Heart Rate")

ax2.legend(*scatter2.legend_elements(), title="Target")
#add horizontal line

ax2.axhline(y=over_50["thalach"].mean(),
          linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

## Customizing matplotlib plots and getting stylish

```python
plt.style.available
```

```python
#set style
plt.style.use("seaborn-muted")

#Using pure OOP
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")
#customizing
ax.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol level")

ax.legend(*scatter.legend_elements(), title="Target")
#add horizontal line

ax.axhline(y=over_50["chol"].mean(),
          linestyle="--");
```
